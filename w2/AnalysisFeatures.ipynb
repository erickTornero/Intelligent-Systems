```python
import pandas as pd
# if you want to download the original file:
#df = pd.read_csv('https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/50k_imdb_movie_reviews.csv')
# otherwise load local file
df = pd.read_csv('shuffled_movie_data.csv')
df.tail()
```

```python
# Return a lower case proccesed text
def processtextT(texto):
    import re
    REPLACE_NO_SPACE = re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])|(\*)(\n)")
    REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")
    texto = REPLACE_NO_SPACE.sub('', texto.lower())
    texto = REPLACE_WITH_SPACE.sub(' ', texto)
    return texto
```

```python
df2 = df.copy()
```

```python
dataframe = pd.DataFrame(columns=['review', 'sentiment'])
```

```python
dataframe
```

```python
for i, row in df2.iterrows():
    dataframe = dataframe.append({'review':processtextT(row['review']), 'sentiment':row['sentiment']}, ignore_index=True)
```

```python
dataframe.head(10)
```

```python
dataframe.to_csv('reviews_processed.csv')
```

# Make data frequency

```python
dictionary = {}
```

```python
import re
for i, row in dataframe.iterrows():
    texto = re.split(' ', row['review'])
    for word in texto:
        if len(word) > 0:
            try:
                if row['sentiment'] == 1:
                    count = dictionary[word][1]
                    dictionary[word][1] = count +1
                elif row['sentiment'] == 0:
                    count = dictionary[word][0]
                    dictionary[word][0] = count + 1
            except KeyError:
                if row['sentiment'] == 1:
                    dictionary[word] = [0, 1]
                elif row['sentiment'] == 0:
                    dictionary[word] = [1, 0]
```

# Filter non valuable words

```python
discriminatory_dictionary = dictionary.copy()
```

### Filter the dictionary

* First filter sum < 50

```python
print('Initial Lenght>', len(discriminatory_dictionary))
for key in dictionary:
    L = dictionary[key]
    suma = L[0]+L[1]
    if suma < 50:
        del discriminatory_dictionary[key]
        
print('Final Lenght>', len(discriminatory_dictionary))
```

* Second filter: percentege between 0.40 and 0.60

```python
# Update the iterative dictionary
dictionary = discriminatory_dictionary.copy()
```

```python
print('Initial Lenght>', len(discriminatory_dictionary))
for key in dictionary:
    L = dictionary[key]
    suma = L[0]+L[1]
    percent = L[0]/suma
    if percent < 0.6 and percent > 0.40:
        del discriminatory_dictionary[key]
        
print('Final Lenght>', len(discriminatory_dictionary))
```

# Creating the dataframe

```python
dataf = pd.DataFrame(columns=['word', 'probneg','probpos','freq'])
```

```python
for key in discriminatory_dictionary:
    L = discriminatory_dictionary[key]
    suma = L[0]+L[1]
    dataf = dataf.append({'word':key, 'probneg':L[0]/suma, 'probpos':L[1]/suma,'freq':L[0]+L[1]}, ignore_index=True)

dataf.index.name = 'Id'
```

```python
dataf.to_csv('frequency_probs.csv')
```

```python
dataf.head(20)
```

# Get the positive features

```python
PosFeatures = dataf.sort_values(['probpos', 'freq', 'probneg'], ascending=[0,0,1])
PosFeatures.reset_index(inplace = True, drop = True)
PosFeatures.index.name = 'Id'
PosFeatures.head(15)
```

```python
posProbArr = PosFeatures['probpos'].values
posFreqArr = PosFeatures['freq'].values
L = list(map(lambda x, y: x*y, posProbArr,posFreqArr))

len(L)
tempDF = pd.DataFrame(data={'PosValue':L})

PosFeaturesExtend = PosFeatures.join(tempDF)
PosFeaturesExtend = PosFeaturesExtend.sort_values(['PosValue'], ascending=False)
PosFeaturesExtend.reset_index(inplace = True, drop = True)
PosFeaturesExtend.index.name = 'Id'
PosFeaturesExtend.head(15)
```

```python
PosFeatures.to_csv('Extract_positive_Features.csv')
PosFeaturesExtend.to_csv('Extract_positive_Features2.csv')
```

# Get the negative features

```python
NegFeatures = dataf.sort_values(['probneg'], ascending=False)
NegFeatures.reset_index(inplace = True, drop = True)
NegFeatures.index.name = 'Id'
NegFeatures.head(15)
```

```python
NegFeatures.to_csv('Extract_negative_Features.csv')
```
